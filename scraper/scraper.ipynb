```python
import requests
from bs4 import BeautifulSoup
```

```python
# Automatically get the HTML of the news link
homepage = requests.get('https://www.medicalnewstoday.com/')
homepage_html = homepage.content
# Parse the page content
homesoup = BeautifulSoup(homepage.text, "html.parser")

homesoup
```

```python
# Add "latest news" list to store the news links
latest_news = []
try:
    for div in homesoup.find_all('li', class_='css-1ib8oek'):
        latest_news.append(div.a['href'])
except:
    pass
```

```python
latest_news
```

```python
with open("medical_document.txt", "r") as file:
    f= file.readlines()
```

```python
# Add an empty list to hold our data
scraped_data = []
```

```python
for url in latest_news:
    if url in f:
        continue
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    # Add url to scraped_data
    scraped_data.append(url)
    # Extract the title: 'h1' tag
    title = soup.h1.text
    scraped_data.append(title)
    # Extract the list below the title in bold: ul - strong tag
    for ul in soup.find_all("ul"):
        for strong in ul.find_all("strong"):
            scraped_data.append(strong.text.strip())
            # print(strong.text.strip())
    # As there's no subheadline for the first paragraph,
    # extract the first paragraph under <div class="css-1u22pos"> tag
    for div in soup.find_all("div", class_="css-1u22pos"):
        for p in div.find_all("p"):
            scraped_data.append(p.text.strip())
            # print(p.text.strip())
    # Extract the other subheadlines and their text under: <div class="css-0"> tag
    for div in soup.find_all("div", class_="css-0"):
        for h2 in div.find_all("h2"):
            scraped_data.append(h2.text.strip())
            # print(h2.text.strip())
        for p in div.find_all("p"):
            scraped_data.append(p.text.strip())
            # print(p.text.strip())
```

```python
# Put everything together
scraped_data

# Write the data to a file
with open("sample_static.txt", "a") as filehandle:
    for listitem in scraped_data:
        filehandle.write("%s\n" % listitem)
```
